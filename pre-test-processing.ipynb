```python
import pandas as pd

df = pd.read_excel('./data/results_pre-test-gescoord-raw.xlsx')
```

```python
df
```

# Get rid of Question, Answer, Correct answer columns

```python
df = df.drop(['Question', 'Answer', 'Correct answer'], axis=1)
```

```python
df
```

# Replace '58+116' with 58 and '60+115' with '60'

```python
df['Kaart id'] = df['Kaart id'].replace(['58+116'], '58')
df['Kaart id'] = df['Kaart id'].replace(['60+115'], '60')
df
```

# Replace 'empty' scores with 0 (e.g. Aarts, Kaart id: 67)

```python
df['Score'] = df['Score'].replace([float('nan')], 0)
```

```python
df
```

```python
import math

students = {}

for index, row in df.iterrows():
    student_id = row['Last Name'] + ' ' + row['First Name'] 
    
    if student_id not in students:
        students[student_id] = { 'User id' : 0, 'Label' : 0, 'First Name' : '', 'Last Name' : '', 'Kaarten' : {} }
        
    if math.isnan(row['User id']) is False:
        students[student_id]['User id'] = int(row['User id'])

    if math.isnan(row['Label']) is False:
        students[student_id]['Label'] = int(row['Label'])

    students[student_id]['First Name'] = row['First Name']
    students[student_id]['Last Name'] = row['Last Name']
    students[student_id]['Kaarten'][int(row['Kaart id'])] = row['Score']
       
len(students)
s = list(students.items())[0]
```

```python
columns = ['Last Name', 'First Name', 'Label']

columns = columns + sorted(s[1]['Kaarten'].keys())
df = pd.DataFrame(columns=columns)
df
```

```python
# fill it up

for key, value in students.items():
    loc = value['User id']
    
    df.loc[loc] = ''
    for key in value.keys():
        if key != 'Kaarten':
            df.loc[loc][key] = value[key]
        else:
            for kaart in value['Kaarten']:
                df.loc[loc][kaart] = value['Kaarten'][kaart]
                
    loc = loc + 1
```
